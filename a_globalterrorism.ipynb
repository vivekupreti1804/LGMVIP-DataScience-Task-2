## Load the data

## Select the useful features or columns

```python
df = df[['iyear','imonth','iday','country_txt','region_txt','provstate','city','latitude','longitude','location','summary','attacktype1_txt','targtype1_txt','gname','motive','weaptype1_txt','nkill','nwound','addnotes']]
df.head()
```

## Rename the Feature name 

```python
df.rename(columns={'iyear':'Year','imonth':"Month",'iday':"Day",'country_txt':'Country',
                    'region_txt':'Region','provstate':'Province','city':'City',
                   'latitude':'Latitude','longitude':'Longitude','loction':'Location',
                   'summary':'Summary','attacktype1_txt':'Attack','targtype1_txt':'Target Type',
                   'gname':'Group Name','motive':"Motive",'weaptype1_txt':'Weapon Type','nkill':'Killed','nwound':'Wounded','addnotes':'Add Notes'},inplace='True')

df.head()
```

```python
 df.info()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Killed']=df['Killed'].fillna(0)
df['Wounded']=df['Wounded'].fillna(0)
df['Casualty']=df['Killed']+df['Wounded']

df.describe()
```

###### Import Points

1. The Data consists of terrorist activites from year 1970 to 2017
2. Maximum number of people killed in an event 1570
3. Maximum number of people wounded in an event 8191
4. Maximum number of total casualities an an event 9574


## Visualize The Data



### (1). Year Wise Attacks

1. Attacks in each year

 

```python
attacks = df['Year'].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={'index':'Year','Year':'Attacks'}).set_index('Year')
attacks.head()
```

```python
attacks.plot(kind='bar',color='cornflowerblue',figsize=(15,6),fontsize=13)
plt.title("Attacks Timeline",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.show()
```

(i). 16903 Attacks in 2014  ( Most Numbers )
(ii). 471 Attacks in 1971 ( Least Number )
  
 2. Total Casualities (Killed+Wounded) in each Y

```python
yc = df[["Year","Casualty"]].groupby("Year").sum()
yc.head()
```

```python
yc.plot(kind='bar',color='cornflowerblue',figsize=(15,6))
plt.title("Year wise Casualies",fontsize=13)
plt.xlabel("Years",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualization",fontsize=13)
plt.show()
```

3. Every year Killed

```python
yk=df[['Year','Killed']].groupby("Year").sum()
yk.head()
```

4. Wounded in each region

```python
yw = df[['Year','Wounded']].groupby('Year').sum()
yw.head()
```

```python
figure=plt.figure()
axis0=figure.add_subplot(2,1,1)
axis1=figure.add_subplot(2,1,2)

#Killed
yk.plot(kind='bar',color='cornflowerblue',figsize=(15,15),ax=axis0)
axis0.set_title("People Killed in each Year")
axis0.set_xlabel("Years")
axis0.set_ylabel("Number of people Killed")

#wounded
yw.plot(kind='bar',color='cornflowerblue',figsize=(15,15),ax=axis1)
axis1.set_title("People Wounded in each year")
axis1.set_xlabel("Years")
axis1.set_ylabel("Number of people Wounded")

plt.show()
```

### (2) Region WIse Attacks
1. Distribution Terrorist Attacks over Regions from 1970-2017


```python
reg = pd.crosstab(df.Year,df.Region)
reg.head()
```

```python
reg.plot(kind='area',stacked=False,alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of attacks",fontsize=20)
plt.show()
```

2. Total Terrorist Attacks in each Region from 1970-2017


```python
regt=reg.transpose()
regt['Total']=regt.sum(axis=1)
ra=regt['Total'].sort_values(ascending=False)
ra
```

```python
ra.plot(kind='bar',figsize=(15,6))
plt.title("Total number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

3. Total Casuality(Killed*wounded) in each Region

```python
rc=df[['Region','Casualty']].groupby("Region").sum().sort_values(by='Casualty',ascending=False)
rc
```

```python
rc.plot(kind='bar',color='cornflowerblue',figsize=(15,6))
plt.title("Region wise Casualties",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

4. Killed in each Region

```python
rk=df[['Region','Killed']].groupby('Region').sum().sort_values(by='Killed',ascending=True)
rk
```

5. Wounded in each Region

```python
rw=df[['Region','Wounded']].groupby("Region").sum().sort_values(by='Wounded',ascending=False)
rw
```

```python
fig=plt.figure()
axis0=fig.add_subplot(1,2,1)
axis1=fig.add_subplot(1,2,2)

#killed
rk.plot(kind='bar',color='cornflowerblue',figsize=(15,6),ax=axis0)
axis0.set_title("People Killed in each Region")
axis0.set_xlabel("Regions")
axis0.set_ylabel("Number of People Killed")

#wounded
rw.plot(kind='bar',color='cornflowerblue',figsize=(15,15),ax=axis1)
axis1.set_title("People Wounded in each Region")
axis1.set_xlabel("Regions")
axis1.set_ylabel("Number of people Wounded")

plt.show()
```
